## Joining the spc districts with all routes

To get the hang out data munging from GIS to python, we find all the columns associatedw with the routes dataset:

```python
routes_paac = os.path.join(workspace2, "Transit_PAAC_All_Stops")
route_col = arcpy.ListFields(routes_paac)

for field in route_col:
    print("{0} is a type of {1} with a length of {2}"
          .format(field.name, field.type, field.length))
```

We want to take the fields for ROUTE:

```python
cursor = arcpy.da.SearchCursor(routes_paac, ['PAAC_Stops_Combined_Data_csv_Routes_ser', 
                                             'PAAC_Stops_Combined_Data_csv_StopID',
                                             'PAAC_Stops_Combined_Data_csv_Stop_name'])
for row in cursor:
    print(row)
```

```python
spcboundaries = os.path.join(workspace2, "Boundaries_SPC_Districts_2020")
spc20_col = arcpy.ListFields(spcboundaries)

for field in spc20_col:
    print("{0} is a type of {1} with a length of {2}"
          .format(field.name, field.type, field.length))
```

```python
# Output will be the target features, states, with a mean city population field (mcp)
outspc = os.path.join(workspace1, "routes_spc_joined")
 
# Create a new fieldmappings and add the two input feature classes.
fieldmappings = arcpy.FieldMappings()
fieldmappings.addTable(routes_paac)
fieldmappings.addTable(spcboundaries)

# Identify fields
keepers = ['PAAC_Stops_Combined_Data_csv_Routes_ser','PAAC_Stops_Combined_Data_csv_StopID',
           'PAAC_Stops_Combined_Data_csv_Stop_name', 'District', 'Name']

# Remove all output fields you don't want.
for field in fieldmappings.fields:
    if field.name not in keepers:
        fieldmappings.removeFieldMap(fieldmappings.findFieldMapIndex(field.name))

#Run the Spatial Join tool, using the defaults for the join operation and join type
arcpy.SpatialJoin_analysis(routes_paac, spcboundaries, outspc,
  join_operation = "JOIN_ONE_TO_ONE",
  join_type = "KEEP_ALL",
  match_option="CLOSEST", field_mapping=fieldmappings)
```

```python
spc_joined = os.path.join(workspace1, "routes_spc_joined")
spc_new_col = arcpy.ListFields(spc_joined)

for field in spc_new_col:
    print("{0} is a type of {1} with a length of {2}"
          .format(field.name, field.type, field.length))
```

```python
cursor = arcpy.da.SearchCursor(spc_joined, [field.name for field in spc_new_col])
for row in cursor:
    print(row)
```
