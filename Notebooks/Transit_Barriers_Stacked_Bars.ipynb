This notebook will compare certain metrics such as barriers to transit and current transit used across a cross section of demographics including race, gender, employment status and age.

The comparison method will be stacked bar charts, which show total shares of the different metrics.

We will look at the overall network as well as individual corridors.

```python
import helpers as hp
import pandas as pd
import numpy as np
```

```python
data = hp.firstNormal(pd.read_csv("../Data/responses_anonymous.csv", 
                                  encoding = "ISO-8859-1", usecols=[6, 11, 12, 13, 14, 15, 17, 18]))
```

```python
[x for x in enumerate(data.columns)]
```

```python
age_ind = 4
age_var = "Age"
emp_ind = 5
emp_var = "Employment_status"
gender_ind = 6
gender_var = "Gender"
race_ind = 7
race_var = "Race"
```
