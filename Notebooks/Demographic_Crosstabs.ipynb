```python
# First will do Ages, Gender, Employment and Race vs Barriers to Transit

data = hp.firstNormal(pd.read_csv("../Data/responses_anonymous.csv", 
                                  encoding = "ISO-8859-1", usecols=[6, 11, 13, 14, 15, 17, 18]))
data.head()
```

## Beginning with just the 67 Corridor

```python
c67 = data[data.Corridor_preference == 'Monroeville via Wilkinsburg (67 Corridor)']
c67.head()
```

### Looking at Barriers to Transit by Demographics (Age, Employment, Gender and Race)

```python
# Age

pd.crosstab(c67.Age, c67.Why_Other_Transport_Used)
```

```python
# Employment Status
pd.crosstab(c67.Employment_status, c67.Why_Other_Transport_Used)
```

```python
# Gender
pd.crosstab(c67.Gender, c67.Why_Other_Transport_Used)
```

```python
# Race
pd.crosstab(c67.Race, c67.Why_Other_Transport_Used)
```
