```python
# First will do Ages, Gender, Employment and Race vs Barriers to Transit

data = hp.firstNormal(pd.read_csv("../Data/responses_anonymous.csv", 
                                  encoding = "ISO-8859-1", usecols=[6, 11, 12, 13, 14, 15, 17, 18]))
data.head()
```

## Beginning with the 67 Corridor

```python
c67 = data[data.Corridor_preference == 'Monroeville via Wilkinsburg (67 Corridor)']
c67.head()
```

### Looking at Barriers to Transit by Demographics (Age, Employment, Gender and Race)

```python
# Age

a = pd.crosstab(c67.Age, c67.Why_Other_Transport_Used)
a.plot.barh(stacked=True)
plt.legend(title='Reasons WHy Other Transport Used', bbox_to_anchor=(1, 2), loc='upper right', ncol=1)
plt.show()
```

```python
# Employment Status
a = pd.crosstab(c67.Employment_status, c67.Why_Other_Transport_Used)
a.plot.barh(stacked=True)
plt.legend(title='Reasons WHy Other Transport Used', bbox_to_anchor=(1, 2), loc='upper right', ncol=1)
plt.show()
```

```python
# Gender
a = pd.crosstab(c67.Gender, c67.Why_Other_Transport_Used)
a.plot.barh(stacked=True)
plt.legend(title='Reasons WHy Other Transport Used', bbox_to_anchor=(1, 2), loc='upper right', ncol=1)
plt.show()
```

```python
# Race
a = pd.crosstab(c67.Race, c67.Why_Other_Transport_Used)
a.plot.barh(stacked=True)
plt.legend(title='Reasons WHy Other Transport Used', bbox_to_anchor=(1, 2), loc='upper right', ncol=1)
plt.show()
```
